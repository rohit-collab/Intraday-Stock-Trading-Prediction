# Let us convert Datatype of date into datetime

```python
data_new=data #keep the original data uneffected, we will copy into new variable
data_new.head()
```

```python
data_new['Date']=pd.to_datetime(data_new['Date'])
data_new.head()
```

```python
data_new.info()
```

# we can observe date datatype is changed to datetime

```python
data_new['Day']=data_new['Date'].dt.day
data_new['Month']=data_new['Date'].dt.month
data_new['Year']=data_new['Date'].dt.year
data_new.head()
```

```python
data_new.info()
```

# CALCULATING DAILY RETURNS
## First, we will create the column ‘Daily Lag’ which is basically just shifting the ‘Close’ price by one day back

```python
data_new['Daily Lag'] = data_new['Close'].shift(1)
data_new.head()
```

```python
data_new['Daily Returns'] = (data_new['Daily Lag']/data_new['Close']) -1
data_new.head()
```

```python
mean = data_new['Daily Returns'].mean()
std = data_new['Daily Returns'].std()
print('Mean =',mean)
print('Standard deviation =',std)
```

```python
plt.figure(figsize=(15,6))
data_new['Daily Returns'].hist(bins=20)
plt.axvline(mean,color='red', linestyle='dashed', linewidth=2)
#to plot the std line we plot both the positive and negative values 
plt.axvline(std,color='g',linestyle='dashed',linewidth=2)
plt.axvline(-std,color='g',linestyle='dashed',linewidth=2)
```

# Let us scale the dataset so that the values fall in the range 0 to 1

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
data_new['Open']=mm.fit_transform(data_new['Open'].values.reshape(-1,1))
data_new['High']=mm.fit_transform(data_new['High'].values.reshape(-1,1))
data_new['Low']=mm.fit_transform(data_new['Low'].values.reshape(-1,1))
data_new['Close']=mm.fit_transform(data_new['Close'].values.reshape(-1,1))
data_new['Adj Close']=mm.fit_transform(data_new['Adj Close'].values.reshape(-1,1))
data_new['Volume']=mm.fit_transform(data_new['Volume'].values.reshape(-1,1))
data_new['Daily Lag']=mm.fit_transform(data_new['Daily Lag'].values.reshape(-1,1))
data_new['Daily Returns']=mm.fit_transform(data_new['Daily Returns'].values.reshape(-1,1))
```
